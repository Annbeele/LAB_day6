```python
#Looks like noise:
X=X.drop(['number_of_open_complaints'],axis=1)
```

```python
X.shape
```

```python
#if I excluded the outliers, I would get:

iqr=np.percentile(X['customer_lifetime_value'],75)-np.percentile(X['customer_lifetime_value'],25)
upper_limit=np.percentile(X['customer_lifetime_value'],75)+(1.5*iqr)
lower_limit=np.percentile(X['customer_lifetime_value'],25)-(1.5*iqr)
excl_outliers=X[(X['customer_lifetime_value']>lower_limit)&(X['customer_lifetime_value']<upper_limit)]

```

```python
sns.boxplot(x=excl_outliers['customer_lifetime_value'])
```

```python
excl_outliers['customer_lifetime_value'].hist()
```

```python
#Try to rescale with log:

def log_transform_clean(x):
    x=np.log(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN

X['customer_lifetime_value']=list(map(log_transform_clean,X['customer_lifetime_value']))
```

```python
X['customer_lifetime_value'].hist()
```
